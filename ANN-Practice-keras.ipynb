```python
from sklearn.preprocessing import LabelEncoder #LabelEncoder assigns numbers to categories ie.enumerates labels
```

```python
# each category value is converted into a new column & assigned a 1 or 0
# (notation for true/false) value to the column
from sklearn.preprocessing import OneHotEncoder
#from sklearn.compose import ColumnTransformer   # include this if (scikit version)>0.20
import numpy as np
```

```python
#ct = ColumnTransformer([("StudentScore", OneHotEncoder(), [1])], remainder = 'passthrough')
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
#X = ct.fit_transform(X) 
```

```python
pd.DataFrame(X,index=X[:,0]).head(2)
```

```python
X = X[:, 1:]
pd.DataFrame(X,index=X[:,0]).head()
```

```python
y=df_all_student.iloc[:, 12]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# StandardScaler is used to resize the distribution of values so that the mean of observed 
# values is 0 and the standard deviation is 1
X_train = sc.fit_transform(X_train) #Fit method calculates mean & variance of each features..
#..present in our data
# fit_transform() is used on the training data so that we can scale the training data & also 
# learn the scaling parameters of that data
X_test = sc.transform(X_test)
```

```python
from tensorflow import keras  # calling keras api to process inputs(i/p) & get outputs(o/p)
from keras.models import Sequential  #stack of layers;layers have exactly 1 i/p & o/p tensor
from keras.layers import Dense #Dense implements element-wise (input*weights + bias) functn
from keras.layers import Dropout #Dropout randomly sets i/p units to 0, avoiding overfitting
```

```python
cf = Sequential()
```

```python
cf.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
cf.add(Dropout(rate = 0.1))
```

```python
cf.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
cf.add(Dropout(rate = 0.1))
cf.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# from ann_visualizer.visualize import ann_viz;
# ann_viz(cf, title=" Simple ");
```

```python
cf.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
X_train.shape
```

```python
cf.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_prediction =cf.predict(X_test)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
def kera_classifier():
    cf = Sequential()
    cf.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    cf.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    cf.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    cf.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return cf
cf = KerasClassifier(build_fn = kera_classifier, batch_size = 10, epochs = 100)
accuracies = cross_val_score(estimator = cf, X = X_train, y = y_train, cv = 10, n_jobs = -1)
mean = accuracies.mean()
variance = accuracies.std()
```
